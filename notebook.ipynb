```python
import numpy as np

from sklearn.model_selection import train_test_split

import matplotlib.pyplot as plt

from sklearn.metrics import confusion_matrix

from sklearn.manifold import TSNE

import seaborn as sns

import librosa

from fefa import (
    load_dataset,
    extract_features,
    train_ubm,
    extract_speaker_supervectors,
    factor_analysis,
    extract_speaker_factors,
    train,
    test,
    cosine_score,
)
```

```python
audios, y = load_dataset()

print(len(audios))
```

```python
plt.plot(5, 5)

speaker_counts = np.bincount(y)[1:]

plt.bar(range(1, 41), speaker_counts)

plt.xlabel("Speaker ID")

plt.ylabel("Number of Samples")

plt.tight_layout()
plt.show()
```

```python
features = extract_features(audios)
```

```python
plt.figure(figsize=(10, 4))
librosa.display.specshow(features[0][:13], x_axis="time", sr=16000)
plt.colorbar(format="%+2.0f dB")
plt.title("MFCC Features")
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 4))
librosa.display.specshow(features[0][13:], x_axis="time", sr=16000)
plt.colorbar(format="%+2.0f dB")
plt.title("Delta MFCC Features")
plt.tight_layout()
plt.show()
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    features, y, test_size=0.2, stratify=y, random_state=42
)

print(len(x_train), len(x_test))
```

```python
ubm = train_ubm(x_train)
```

```python
supervectors_train = extract_speaker_supervectors(x_train, ubm)

print(supervectors_train.shape)
```

```python
fa = factor_analysis(supervectors_train)

print(fa.components_.shape)
```

```python
speaker_factors = extract_speaker_factors(supervectors_train, fa)

print(speaker_factors.shape)
```

```python
model, wccn_matrix = train(speaker_factors, y_train)
```

```python
lda_factors = model.transform(speaker_factors)

tsne = TSNE(n_components=2, random_state=42, perplexity=30)
tsne_factors = tsne.fit_transform(lda_factors)

plt.figure(figsize=(10, 8))

for speaker in np.unique(y_train)[:5]:
    mask = y_train == speaker

    plt.scatter(
        tsne_factors[mask, 0],
        tsne_factors[mask, 1],
        label=f"Speaker {speaker}",
        alpha=0.6,
    )

plt.title("LDA Transformed Speaker Factors")
plt.xlabel("t-SNE component 1")
plt.ylabel("t-SNE component 2")
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
plt.tight_layout()

plt.show()
```

```python
def evaluate_model(x_test, y_test):
    test_supervectors = extract_speaker_supervectors(x_test)
    test_factors = extract_speaker_factors(test_supervectors)

    test_wccn_factors = test(test_factors, model, wccn_matrix)
    train_wccn_factors = test(speaker_factors, model, wccn_matrix)

    unique_speakers = np.unique(y_train)
    speaker_embeddings = []

    for speaker in unique_speakers:
        speaker_mask = np.array(y_train) == speaker
        speaker_mean = np.mean(train_wccn_factors[speaker_mask], axis=0)
        speaker_embeddings.append(speaker_mean)

    speaker_embeddings = np.array(speaker_embeddings)

    scores = cosine_score(speaker_embeddings, test_wccn_factors)

    predictions = unique_speakers[np.argmax(scores, axis=1)]

    accuracy = np.mean(predictions == np.array(y_test))

    return accuracy, predictions


accuracy, predictions = evaluate_model(
    x_test, y_test, x_train, y_train, model, wccn_matrix, speaker_factors
)

print(f"Model Accuracy: {accuracy:.2%}")
```

```python
cm = confusion_matrix(y_test, predictions)

plt.figure(figsize=(15, 15))

sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")

plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.tight_layout()

plt.show()
```
